#### QStN NO.1->How many are there in each Team and the percentage splitting with respect to the total employees.

```python
value=df['Team'].value_counts()
print(value)
value.plot(kind='bar')
```

```python
#for finding the percentage splitting first finding the length of dataframe
length=len(df)
print(length)
```

```python
#Percentage splitting
percentage=value/length*100
print(percentage)
```

```python

# Visual representation with a pie chart
team_percentage = df['Team'].value_counts() / length * 100
plt.figure(figsize=(8, 8))
plt.pie(team_percentage, labels=team_percentage.index, autopct='%1.1f%%', startangle=90, colors=plt.cm.Paired.colors)
plt.title('Percentage of Employees in Each Team')
plt.show()
```

### QSTN NO2:Segregate the employees w.r.t different positions.

```python
#employees wrt different positions
position_counts = df['Position'].value_counts()
print(position_counts)
# Visual representation with a bar chart
plt.figure(figsize=(10, 6))
position_counts.plot(kind='bar', color='green')
plt.title('Number of Employees in Each Position')
plt.xlabel('Position')
plt.ylabel('Count')
plt.xticks(rotation=45, ha='right')
plt.show()
```

### QSTN NO3:Find from which age group most of the employees belong to.

```python
most_common_age=df['Age'].value_counts().idxmax()
print(most_common_age)
```

### QSTNNO4:Find out under which team and position, spending in terms of salary is high.

```python
high_salary_team_position = df.groupby(['Team', 'Position'])['Salary'].sum().idxmax()
print(high_salary_team_position)

```

```python
# Visual representation of salary spending for each team and position
plt.figure(figsize=(35,25))
salary_spending_by_team_position = df.groupby(['Team', 'Position'])['Salary'].sum().sort_values(ascending=False)
salary_spending_by_team_position.plot(kind='bar', color='green')
plt.title('Salary Spending by Team and Position')
plt.xlabel('Team, Position')
plt.ylabel('Total Salary Spending')
plt.xticks(rotation=45, ha='right')
plt.show()

```

### QSTN NO5:Find if there is any correlation between age and salary , represent it visually.

```python
#here finding the correlation matrix of age and salary used by crr() mathod
#after the correlation ,it represent visually help with Heatmap
subset_of_age_salary=df[['Age','Salary']]
corr=subset_of_age_salary.corr()
#create heatmap
sns.heatmap(corr,annot=True,cmap="viridis")
# the title of the gragh
plt.title("Correlation between Age and Salary")
#Display the heatmapj
plt.show()
```
